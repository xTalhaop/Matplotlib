```python
import matplotlib.pyplot as plt
import numpy as np

xp= np.array([0,10])
yp= np.array([0,125])
plt.plot(xp,yp)
# Show the plot
plt.show()
```

```python
import matplotlib.pyplot as plt

plt.plot([1, 2, 3], [4, 5, 6], label='Line 1')  # Add a label
plt.plot([1, 2, 3], [6, 5, 4], label='Line 2')  # Add another label

plt.legend()  # Display the legend
plt.show()

```

```python
import matplotlib.pyplot as plt

# Data to plot
sizes = [3, 20, 25, 25]                  # The values for each slice
labels = ['Apples', 'Bananas', 'Cherries', 'Dates']  # Labels for each slice

# Create pie chart
plt.pie(sizes, labels=labels,colors = ['red', 'yellow', 'purple', 'brown'], autopct='%1.1f%%', startangle=0)

plt.title('Fruit Distribution')
plt.axis('equal')  # Ensures the pie chart is a circle
plt.show()
```

```python
import matplotlib.pyplot as plt

# Data
categories = ['A', 'B', 'C', 'D']
values = [10, 24, 36, 18]

# Create bar graph
plt.bar(categories, values, color=['red', 'blue', 'green', 'orange'])
# Add labels and title
plt.xlabel('Category')
plt.ylabel('Value')
plt.title('Simple Bar Graph')

# Show the plot
plt.show()
```

```python
import numpy as np

x = np.arange(4)
y = np.random.rand(4)
values = [10, 20, 30, 400]

plt.scatter(x, y, c=values, cmap='viridis')  # Color by values
plt.colorbar()  # Show color scale
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Generate mock data: 1000 random numbers from a normal distribution
data = np.random.randn(1000)

# Create histogram
plt.hist(data, bins=30, color='skyblue', edgecolor='black')

# Add titles and labels
plt.title('Mock Histogram of Normally Distributed Data')
plt.xlabel('Value')
plt.ylabel('Frequency')

# Show the plot
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Load an example dataset
tips = sns.load_dataset("tips")

# Create a scatterplot with hue
sns.scatterplot(x="total_bill", y="tip", hue="time", data=tips)
plt.title("Tips by Total Bill, colored by Time of Day")
plt.show()
```
